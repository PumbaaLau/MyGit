```python
import itertools
```

```python
natuals = itertools.count(1)
```

```python
for n in natuals:
    print(n)
```

```python
natuals = itertools.count(1)
```

```python
ns = itertools.takewhile(lambda x: x <= 10, natuals)
```

```python
list(ns)
```

## chain()

chain()

```python
for c in itertools.chain('ABC', 'XYZ'):
    print(c)
```

## groupby()

groupby()

```python
for key, group in itertools.groupby('AAABBBCCAAA'):
    print(key, list(group))
```
