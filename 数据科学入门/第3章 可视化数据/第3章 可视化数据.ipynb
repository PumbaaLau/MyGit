数据可视化有两种主要用途：
- 探索数据

## 3.1 matplotlib

```python
import matplotlib.pyplot as plt
```

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# 创建一幅线图，x轴是年份，y轴是gdp
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')

# 添加一个标题
plt.title("GDP")

# 给y轴加标记
plt.ylabel("Billion")
plt.show()
```

## 3.2 条形图

```python
movies = ['Annie Hall', 'Ben-Hur', 'Casablanca', 'Gandhi', 'West Side Story']
num_oscars = [5, 11, 3, 8, 10]

# 条形的默认宽度是0.8，因此我们对左侧坐标加上0.1
# 这样每个条形就被放置在中心了
xs = [i+0.1 for i,_ in enumerate(movies)]

# 使用左侧x坐标[xs]和高度[num_oscars]画条形图
plt.bar(xs, num_oscars)

plt.ylabel("Number of Oscars")
plt.title("My favourite move")

# 使用电影的名字标记x轴，位置在x轴上条形的中心
plt.xticks([i+0.5 for i,_ in enumerate(movies)], movies)

plt.show()
```

## 3.3 线图

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = sorted(variance, reverse=True)
total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i,_ in enumerate(variance)]

# 可以多次调用plt.plot以便在同一个图上显示多个序列
plt.plot(xs, variance, 'g-', label='variance') # 绿色实线
plt.plot(xs, bias_squared, 'r-.', label='bias^2') # 红色点虚线
plt.plot(xs, total_error, 'b:', label='total error') # 蓝色点线

# 因为已经对每个序列都指派了标记，所以可以自由地不知图例
# loc=9指的是“顶部中央”
plt.legend(loc=9)
plt.xlabel("Complexity of model")
plt.title("bias-variance graph")
plt.show()
```

## 3.4 散点图

```python
friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = list('abcdefghi')

plt.scatter(friends, minutes)

# 每个点加标记
for label, friends_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label, 
                xy = (friends_count, minute_count), # 把标记放在对应的点上
                xytext = (5, -5), # 但要有轻微偏离
                textcoords = 'offset points')
    
plt.title('Number minutes,friends')
plt.xlabel('Number of friends')
plt.ylabel('Minutes cost on web')
plt.show()
```

```python
test_1_grades = [99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]

plt.scatter(test_1_grades, test_2_grades)
plt.title("Axes Aren't Comparable")
plt.xlabel('test_1_grades')
plt.ylabel('test_2_grades')
plt.show()
```
